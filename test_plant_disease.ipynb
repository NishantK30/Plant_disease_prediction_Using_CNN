## Importing libraries

```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
```

### Loading model

```python
model = tf.keras.models.load_model("models\Trained_model_0.1.1.keras")
```

### Visualizing Images of the Test set

```python
import cv2
image_path = "test/TomatoYellowCurlVirus2.JPG"
img = cv2.imread(image_path)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.imshow(img)
plt.title("Test Image")
plt.xticks([])
plt.yticks([])
plt.show()
```

## Testing model

### preprocessing the test image

```python
image = tf.keras.preprocessing.image.load_img(image_path,target_size=(128,128))
input_arr = tf.keras.preprocessing.image.img_to_array(image)
input_arr = np.array([input_arr]) #Conver singlw image to a batch
print(input_arr.shape)
```

```python
prediction = model.predict(input_arr)
```
